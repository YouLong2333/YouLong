```python
def text_preprocessing(typhoon_name):
    tweet=[]
    with open(typhoon_name + '.csv','r') as f:
        next(f)
        reader = csv.reader(f)
        for row in reader:
            tweet.append(row[3]) # append 3rd col in csv
    my_string = " ".join(str(x) for x in tweet)
    my_cleanedString = clean_text(my_string)
    return my_cleanedString
```

```python
## text cleaning --- very important!
def clean_text(text):
    # print(text)

    ##update nltk database
    # import nltk
    # nltk.download()

    text = text.replace("b'", " ")#important

    # split into words
    from nltk.tokenize import word_tokenize
    tokens = word_tokenize(text)
    # convert to lower case
    tokens = [w.lower() for w in tokens]
    # remove punctuation from each word
    import string
    table = str.maketrans('', '', string.punctuation)      #trantab2 = str.maketrans(intab,outtab,deltab) #创建字符映射转换表，并删除指定字符
    stripped = [w.translate(table) for w in tokens]     #print(test.translate(trantab2))
    # remove remaining tokens that are not alphabetic
    words = [word for word in stripped if word.isalpha()] #isalpha() 方法检测字符串是否只由字母组成
    # filter out stop words

    from nltk.corpus import stopwords
    stop_words = set(stopwords.words('english'))
    stop_words.add('b')
    stop_words.add('bi')
    stop_words.add('n')
    stop_words.add('uu')
    stop_words.add('de')
    stop_words.add('https')
    stop_words.add('http')
    stop_words.add('typhoon')
    stop_words.add('wwwyoutubecomwatch')
    # stop_words.add('youtube')
    # stop_words.add('playlist')
    words = [w for w in words if not w in stop_words]
    # print(words[:100])
    # print(words)

    return ' '.join(words)
```

```python
def nltk_sentiment(sentence):
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    
    nltk_sentiment = SentimentIntensityAnalyzer()
    score = nltk_sentiment.polarity_scores(sentence)
    return score
```
